## load Haar cascade for face detection

```python
face_classifier = cv2.CascadeClassifier(r'C:\Users\surav\OneDrive\Desktop\face detection from img pro-1\haarcascade_frontalface_default.xml')
```

## load the image

```python
image = cv2.imread(r'C:\Users\surav\OneDrive\Desktop\face detection from img pro-1\Tom_Holland_MTV_2018_(02).jpg')
```

## check image is loaded or not

```python
if image is None:
    print('Error:Image not found or cannot load')
    exit()
```

## convert image to grayscale

```python
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

## detect the faces in the image

```python
faces = face_classifier.detectMultiScale(gray,1.3,5)
```

## check if faces are detected

```python
if len(faces) == 0:
    print("No faces found!")
else:
    # Draw rectangles around the faces
    for (x, y, w, h) in faces:  # (x, y) is the top-left corner, and (w, h) is the width and height of the face
        cv2.rectangle(image, (x, y), (x + w, y + h), (127, 0, 255), 2)

    # Display the output image
    cv2.imshow('Face Detection', image)
    cv2.waitKey(0) 
```

```python
import cv2
import gradio as gr
import numpy as np

# Load Haar Cascade (keep XML in same folder or give correct path)
face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)

def detect_faces(image):
    """
    image: PIL Image from Gradio
    return: image with detected faces
    """
    # Convert PIL image to OpenCV format
    image = np.array(image)
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    faces = face_cascade.detectMultiScale(
        gray,
        scaleFactor=1.3,
        minNeighbors=5
    )

    if len(faces) == 0:
        return image, "No faces detected ‚ùå"

    for (x, y, w, h) in faces:
        cv2.rectangle(
            image,
            (x, y),
            (x + w, y + h),
            (127, 0, 255),
            2
        )

    return image, f"Faces detected: {len(faces)} ‚úÖ"


# Gradio UI
interface = gr.Interface(
    fn=detect_faces,
    inputs=gr.Image(type="pil", label="Upload Image"),
    outputs=[
        gr.Image(type="numpy", label="Detected Faces"),
        gr.Textbox(label="Detection Result")
    ],
    title="üßë Face Detection using Haar Cascade",
    description="Upload an image and detect faces using OpenCV Haar Cascade.",
    theme="soft"
)

if __name__ == "__main__":
    interface.launch()
```
